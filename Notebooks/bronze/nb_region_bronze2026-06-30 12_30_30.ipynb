```python
env = dbutils.widgets.get("environment") if dbutils.widgets.getAll().get("environment") else "dev"
targetTable = f"saleslake_{env}.bronze_{env}.rawRegion"
srcPath = f"/Volumes/saleslake_{env}/bronze_{env}/vol_saleslake_src_files_{env}/daily_region/"

spark.sql(f"""
    COPY INTO {targetTable}
    FROM (
        SELECT
            region_id,
            region_code,
            region_name,
            country,
            sub_region,
            regional_manager,
            created_date,
            current_timestamp() AS ingest_ts
        FROM '{srcPath}'
    )
    FILEFORMAT = CSV
    FORMAT_OPTIONS ('header' = 'true')
    COPY_OPTIONS ('mergeSchema' = 'false')
    """)

print("Region Bronze Load Successful")
```
