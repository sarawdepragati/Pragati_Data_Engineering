```python
env = dbutils.widgets.get("environment") if dbutils.widgets.getAll().get("environment") else "dev"
targetTable = f"saleslake_{env}.bronze_{env}.rawStore"
srcPath = f"/Volumes/saleslake_{env}/bronze_{env}/vol_saleslake_src_files_{env}/daily_store/"

spark.sql(f"""
    COPY INTO {targetTable}
    FROM (
        SELECT
            store_id,
            store_code,
            store_name,
            store_type,
            address,
            city,
            state,
            region_id,
            manager_name,
            opening_date,
            square_feet,
            status,
            current_timestamp() AS ingest_ts
        FROM '{srcPath}'
    )
    FILEFORMAT = CSV
    FORMAT_OPTIONS ('header' = 'true')
    COPY_OPTIONS ('mergeSchema' = 'false')
    """)

print("Store Bronze Load Successful")
```
