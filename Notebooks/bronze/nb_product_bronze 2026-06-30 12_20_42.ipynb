```python
env = dbutils.widgets.get("environment") if dbutils.widgets.getAll().get("environment") else "dev"
targetTable = f"saleslake_{env}.bronze_{env}.rawProduct"
srcPath = f"/Volumes/saleslake_{env}/bronze_{env}/vol_saleslake_src_files_{env}/daily_product/"

spark.sql(f"""
    COPY INTO {targetTable}
    FROM (
        SELECT
            product_id,
            sku,
            product_name,
            category,
            sub_category,
            brand,
            supplier,
            unit_cost,
            list_price,
            launch_date,
            status,
            current_timestamp() AS ingest_ts
        FROM '{srcPath}'
    )
    FILEFORMAT = CSV
    FORMAT_OPTIONS ('header' = 'true')
    COPY_OPTIONS ('mergeSchema' = 'false')
    """)

print("Product Bronze Load Successful")
```
